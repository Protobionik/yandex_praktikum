````python
current_period = current_period.query('platform != "PS3" and platform != "X360"')#исключу неактуальные платформы из таблицы
````

**Рассмотрим как влиют на продажи отзыва пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами.**

````python
score_ps4 = current_period.query('platform == "PS4"')
fig, ax = plt.subplots(figsize=(9, 5))
ax.scatter(x = score_ps4['total_sales'], y = score_ps4['user_score'])
fig.suptitle('Пользовательская оценка')
plt.show()
user_score = score_ps4['total_sales'].corr(score_ps4['user_score'])
print(f'Корреляция для отзывов пользователей составляет: {user_score:.2f}')
````

````python
score_ps4 = current_period.query('platform == "PS4"')
fig, ax = plt.subplots(figsize=(9, 5))
ax.scatter(x = score_ps4['total_sales'], y = score_ps4['critic_score'])
fig.suptitle('Оценка критиков')
plt.show()
critic_score = score_ps4['total_sales'].corr(score_ps4['critic_score'])
print(f'Корреляция для отзывов критиков составляет: {critic_score:.2f}')
````

**Из этих двух графиков корреляции делаем выводы:**
- Продажи не зависят от отзывов пользователей, пользователи получают игру после выхода ее в релиз. До этого нет возможности оценить продукт. Демо версии мало кто выпускает.
- От оценки критиков зависят продажи, это видно из графика. Так же критики имеют возможность оценить игру до релиза, поэтому решающее слово именно за ними.

**Рассмотрим зависимость от отзывов для остальных платформ**

**Рассмотрев графики корреляции для остальных платформ делаем вывод: оценка критиков опеределенно имеет положительное влияние  на продажи игр, но даже при хороших оценках игры могут плохо продаваться.**

````python
#из комментария ревьюера
import matplotlib.pyplot as plt
df_sc, y = current_period.copy(), 'total_sales'
for platform, games_on_pl in df_sc.groupby('platform'):
    print('='*60)
    print('Расчет по Платформе',platform)
    # Считаем сколько в колонке не пустых отзывов
    not_user = len(games_on_pl[games_on_pl['user_score'].notna() == True])
    not_critic = len(games_on_pl[games_on_pl['critic_score'].notna() == True])
    sum_not = 3 # Задаем количество не пустых значений для вывода диаграммы и расчета корреляции
    if not_user > sum_not:
        games_on_pl.plot(kind='scatter', x='user_score', y=y, color='b')
        display(games_on_pl[['user_score', y]].corr()[y])
        plt.xlabel('Оценки отзывов пользователей')
        plt.ylabel('Продажи в млн.копий')
        plt.title('Диаграмма рассеяния по платформе '+platform)
        plt.show()
    else:
        print('Для платформы',platform, 'не хватает данных для построения диаграммы и расчета корреляции отзывов пользователей')
    if not_critic > sum_not:
        games_on_pl.plot(kind='scatter', x='critic_score', y=y, color='b')
        display(games_on_pl[['critic_score', y]].corr()[y])
        plt.xlabel('Оценки отзывов критиков')
        plt.ylabel('Продажи в млн.копий')
        plt.title('Диаграмма рассеяния по платформе '+platform)
        plt.show()
    else:
        print('Для платформы',platform, 'не хватает данных для построения диаграммы и расчета корреляции отзывов критиков')
````

**Рассмотрев графики и посчитав корреляцию делаев вывод:**
- В большенстве случаев оценка критиков опеределенно имеет положительное влияние на продажи игр, но даже при хороших оценках игры могут плохо продаваться.
- Так же следует заметить что пользовательская оценка редко оказывает значительное влияние на продажи, а в некоторые моменты она имеет очень незначительную отрицательную зависимость.
- Топовые платформы WiiU, PS4, XOne имеют следующие зависимости:
  - PS4 отрицатально слабо зависит от оценок пользователей и слабую положительную зависимость от оценок критиков.
  - XOne не зависит от оценое пользователей и слабо зависит от оценки критиков.
  - WIiU слабо зависит от оценки пользователей, как и слабо на нее влияют отценки критиков.

````python
# Комментарий ревьюера 2
import matplotlib.pyplot as plt
df_sc, y = current_period.copy(), 'total_sales'
for platform, games_on_pl in df_sc.groupby('platform'):
    print('='*60)
    print('Расчет по Платформе',platform)
    # Считаем сколько в колонке не пустых отзывов
    not_user = len(games_on_pl[games_on_pl['user_score'].notna() == True])
    not_critic = len(games_on_pl[games_on_pl['critic_score'].notna() == True])
    sum_not = 3 # Задаем количество не пустых значений для вывода диаграммы и расчета корреляции
    if not_user > sum_not:
        games_on_pl.plot(kind='scatter', x='user_score', y=y, color='b')
        display(games_on_pl[['user_score', y]].corr()[y])
        plt.xlabel('Оценки отзывов пользователей')
        plt.ylabel('Продажи в млн.копий')
        plt.title('Диаграмма рассеяния по платформе '+platform)
        plt.show()
    else:
        print('Для платформы',platform, 'не хватает данных для построения диаграммы и расчета корреляции отзывов пользователей')
    if not_critic > sum_not:
        games_on_pl.plot(kind='scatter', x='critic_score', y=y, color='b')
        display(games_on_pl[['critic_score', y]].corr()[y])
        plt.xlabel('Оценки отзывов критиков')
        plt.ylabel('Продажи в млн.копий')
        plt.title('Диаграмма рассеяния по платформе '+platform)
        plt.show()
    else:
        print('Для платформы',platform, 'не хватает данных для построения диаграммы и расчета корреляции отзывов критиков')
````

**Рассмотрим распределение по жанрам, выберем самые прибыльные**

### Проверка гипотез.

Нужно проверить гипотезу: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

```
H_0: Средний пользовательский рейтинг платформы Xbox One = Среднему пользовательскому рейтингу платформы PC`
H_a: Средний пользовательский рейтинг платформы Xbox One ≠ Среднему пользовательскому рейтингу платформы PC`
alpha = 0.05
```

````python
xone = data[data['platform'] =='XOne']['user_score']
pc = data[data['platform'] =='PC']['user_score']

print(np.var(xone, ddof=1))
print(np.var(pc, ddof=1))
````

````python
results = st.ttest_ind(xone.dropna(), pc.dropna(), equal_var=False )

alpha = 0.05

 
print('p-значение:', results.pvalue)


if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

**Вывод: определенно пользовательский рейтинг разных платформ может быть разным.**

Гипотеза: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
У нас всегда H0 о равенстве чего-то, а H1 о каком-то неравенстве, поэтому алгоритм проверки гипотез всегда абсолютно одинаковый.

```
H_0: Средний пользовательский рейтинг жанра Action = Среднему пользовательскому жанра Sports`
H_a: Средний пользовательский рейтинг жанра Action ≠ Среднему пользовательскому жанра Sports`
alpha = 0.05
```

````python
action = data[data['genre'] =='Action']['user_score']
sports = data[data['genre'] =='Sports']['user_score']

print(np.var(action, ddof=1))
print(np.var(sports, ddof=1))
````

````python
results = st.ttest_ind(action.dropna(), sports.dropna(), equal_var=False )

alpha = 0.05

 
print('p-значение:', results.pvalue)


if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

**Нулевую гипотезу о равенстве жанров Экшн и Спорт отвергнуть не получилось, а значит их пользовательский рейтинг равен.**

**В данном случае можно применить ttest_ind (t-тест), потому что он подходит под основные критерии:**
- Генеральные совокупности не должны зависеть друг от друга. 
- Выборочные средние должны быть нормально распределены.
- Дисперсии рассматриваемых генеральных совокупностей должны быть равны. Дисперсии не равны, ставим параметр equal_var=False.

**В первом случае нужно проверить пользовательский рейтинг по двум платформам, у нас есть две выборки, которые не зависят друг от друга, распределены они нормально и их дисперии не равны, указываем это в параметре т-теста.
Во стором случае задача идентична только перефразирована видимо чтобы ввести в заблуждение, но у нас всегда H0 о равенстве чего-то, а H1 о каком-то неравенстве, поэтому алгоритм проверки гипотез всегда абсолютно одинаковый.**
